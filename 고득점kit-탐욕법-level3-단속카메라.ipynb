````python
def solution(routes):
    
    # 만약 리더중에 해당 차량과 범위 겹치는거 없으면  
    # 리더에 해당 차량 포함시키고 그다음 번부터의 차량의 범위 비교에 활용한다.   
    leaders = []  
    # 다음과 같이 첫키로 솔트하는 조건 넣으니 통과함. 
    routes.sort(key= lambda x:x[0])      
    leaders.append(routes.pop(0))

    
    count = 1 
    for i,car in enumerate(routes):
        
        # 해당 i 차량이 리더중에 겹치는게 있는지 판단.
        picked = False

        for j,item in enumerate(leaders):  
            
            lx,ly = item 
            
            if (lx<=car[0]<=ly) or (lx<=car[1]<=ly):
                picked = True
                
                # 두 범위에 공통 범위로 해당 leader를 줄여줘야함(최신화) 
                leaders[j] = [max(car[0],lx),min(car[1],ly)]
#                 print(car, [lx,ly])
                break
        
        if picked:
            pass 

        else:
            # 없으면 리더에 추가 및 count 
            leaders.append(car)
            count+=1
            
    return count  



# 왜 정렬 조건 필요한지?   
# 그래야만 최초에 뽑힌 leader에 최대한 많이 겹치는 부분 처 넣을수있으니까.
# 정렬안하면 리더로 뽑히는 차량 주변에 동떨어진 친구들 있어서 -> leader를 최신화 해 나갈때 팍팍 줄어든다.   
# 정렬하면 그 leader변동 폭이 한번 마다 적어서 최대한 같은 범위 안에 많이 처 넣을 수 있다.   
````
