문제 설명  
일반적인 프린터는 인쇄 요청이 들어온 순서대로 인쇄합니다. 그렇기 때문에 중요한 문서가 나중에 인쇄될 수 있습니다. 이런 문제를 보완하기 위해 중요도가 높은 문서를 먼저 인쇄하는 프린터를 개발했습니다. 이 새롭게 개발한 프린터는 아래와 같은 방식으로 인쇄 작업을 수행합니다.  
  
1. 인쇄 대기목록의 가장 앞에 있는 문서(J)를 대기목록에서 꺼냅니다.    
2. 나머지 인쇄 대기목록에서 J보다 중요도가 높은 문서가 한 개라도 존재하면 J를 대기목록의 가장 마지막에 넣습니다.  
3. 그렇지 않으면 J를 인쇄합니다. 
예를 들어, 4개의 문서(A, B, C, D)가 순서대로 인쇄 대기목록에 있고 중요도가 2 1 3 2 라면 C D A B 순으로 인쇄하게 됩니다.  
  
내가 인쇄를 요청한 문서가 몇 번째로 인쇄되는지 알고 싶습니다. 위의 예에서 C는 1번째로, A는 3번째로 인쇄됩니다.  
  
현재 대기목록에 있는 문서의 중요도가 순서대로 담긴 배열 priorities와 내가 인쇄를 요청한 문서가 현재 대기목록의 어떤 위치에 있는 지를 알려주는 location이 매개변수로 주어질 때, 내가 인쇄를 요청한 문서가 몇 번째로 인쇄되는지 return 하도록 solution 함수를 작성해주세요.  
  
제한사항  
현재 대기목록에는 1개 이상 100개 이하의 문서가 있습니다.  
인쇄 작업의 중요도는 1~9로 표현하며 숫자가 클수록 중요하다는 뜻입니다.  
location은 0 이상 (현재 대기목록에 있는 작업 수 - 1) 이하의 값을 가지며 대기목록의 가장 앞에 있으면 0, 두 번째에 있으면 1로 표현합니다.  
입출력 예  
priorities	location	return  
[2, 1, 3, 2]	2	1  
[1, 1, 9, 1, 1, 1]	0	5

참고   

맨 첫번째만 maximum priority로 뽑고 나머지는 그냥 뽑겠다는 문제가 아니에요. 모든 task를 priority 순서대로 처리할 것인데, 그 과정에서 밀리는 task들은 모두 다시 맨뒤부터 기다리는 상황을 생각하는거죠.   
예시랑 조금 다르게 [1, 2, 3, 2'] (두번째 2는 구분을 위해 2'이라 합시다) 가 뽑히는 과정을 생각해볼게요  
[1, 2, 3, 2'] - 1은 maximum priority가 아니므로 맨뒤  
[2, 3, 2', 1] - 2 역시 아니므로 맨뒤  
[3, 2', 1, 2] - 3은 max이므로 뽑힘!   
[2', 1, 2] - 2'도 max이므로 뽑힘!  
[1, 2] - 1은 max가 아니므로 맨뒤  
[2, 1] - 2는 max이므로 뽑힘!  
[1] - 1 뽑힘  
이 경우와 본인의 경우가 어떻게 다른지 확인해보세요~  

```python
# 셋시도 
def solution(priorities, location):
    
    counter = 0
    
    # 전략 
    
    # 0. 맨앞에 해당 리스트에서 가장 큰수가 오는경우 -> pop(0) , location 감소 (pop한 횟수 카운트) 
         # 0.1 맨앞에 가장 큰수가 타깃값과 같은경우 -> 해당 locatoin= 0 될때까지 pop하고 그때까지 타깃과 같은값들 센다. 
               # 0.1.1 맨앞이 가장 큰수고 location 0 인경우 -> 지금까지 pop한횟수 + 1 해서 리턴 
        
    # 1. 맨앞이 큰수가 아닌경우 pop(0)한값 뒤로 전달, location 감소 or 맨뒤로 (pop한 횟수 카운트 x) 
        # 1.1 location = 0 아닌경우 -> locaton감소 
        # 1.2 lcation = 0 인경우  -> 맨뒤로 인덱스 변경. 
    
    def check_is_biggest(lista, num):
        return max(lista)==num
    
    # 타깃값 
    target_value = priorities[location]
    
    # 최종 결과 카운터 
    result_counter = 0
    finish_counter = False
    
    while True:
        # print(location)
        
#         counter +=1
#         if counter == 20:
#             break
        
        first_item = priorities[0]
        
        # 맨앞이 가장큰 값이라면 
        if check_is_biggest(priorities, first_item):
            first_item = priorities.pop(0)
            result_counter+=1
            location-=1
            
            # 맨앞의 값이 로케이션 인덱스 0인것이었다면? 
            if location == -1:
                break
            # 맨앞의 값이 타깃 벨류 값과 같다면?(원하는 location 은 맨앞이 아니였을때)
            if target_value == first_item:
                # 현시점 부터 location = 0 되는 순간 까지 타깃값과 같은 값의 갯수 카운트 
                partial_counter = 0
                for item in priorities:  # 바로 현시점 locaiton == 0 이었다면? 
                    
                    if item == target_value:
                        partial_counter+=1
                        
                    if location == 0:
                        result_counter = result_counter + partial_counter
                        finish_counter = True
                        break

                    location-=1
            
    # 1. 맨앞이 최대가 아닌경우 pop(0)한값 뒤로 전달, location 감소 or 맨뒤로 (pop한 횟수 카운트 x) 
        # 1.1 location = 0 아닌경우 -> locaton감소 
        # 1.2 lcation = 0 인경우  -> 맨뒤로 인덱스 변경. 
        
        else:
            first_item = priorities.pop(0)
            priorities.append(first_item)
            if location  == 0:
                location = len(priorities) - 1 
            else:
                location-=1
                
                
        if finish_counter:
            break    
                

        
        
            
        
        
    answer = result_counter
    return answer
```

```python
# 우수답안. 

# 하나의 루프에서 다 처리 가능했다....
# 맨앞이 가장크고 타깃벨류와 같은경우 나눌필요없었다..... 좀더 일반화 해서 생각하자. 
# 루프 두개 나오면 의심부터 해보도록. 한번의루프에서 다 처리하자. 
# 이거처럼 변수도 간결하게 해버려서 한번에 읽을수 있게 해보자. 

def solution(p, l):
    ans = 0
    m = max(p)
    
    while True:
        v = p.pop(0)
        if m == v:
            ans += 1
            if l == 0:
                break
            else:
                l -= 1
            m = max(p)
        else:
            p.append(v)
            if l == 0:
                l = len(p)-1
            else:
                l -= 1
    return ans
```

```python
# priorities	location	return  
# [2, 1, 3, 2]	2	1  
# [1, 1, 9, 1, 1, 1]	0	5  

p = [2, 1, 3, 2]
l = 2 
solution(p, l)
```

```python
# 우수답 대로 다시해본거 

# 우수답안 참고   
def solution(priorities, location):
    
    
    max_value = max(priorities)
    counter = 0
    
    while True:
        first_value = priorities.pop(0)
        
        if max_value == first_value:
            counter +=1
            
            if location == 0:
                break
            
            else:
                location -=1
        
            max_value = max(priorities)
        
        else:
            priorities.append(first_value)
            if location == 0:
                location = len(priorities)-1
            else:
                location -=1
            
        
    
    return counter
    
```
